## Comprehensions: inline loops

Use comprehensions for readable transformations and filters.

Patterns:

```python
[expression for item in iterable]
[expression for item in iterable if condition]
{expression for item in iterable}
{key: value for item in iterable}
(expression for item in iterable)  # generator expression
```

````python
# List comprehension
high_risk_names = [patient["name"] for patient in patients if patient["risk_score"] >= 0.75]
print(high_risk_names)

# Set comprehension
unique_conditions = {patient["condition"] for patient in patients}
print(unique_conditions)

# Dict comprehension
risk_by_patient_id = {patient["patient_id"]: patient["risk_score"] for patient in patients}
print(risk_by_patient_id)

# Generator expression: memory-friendly
average_age = sum(patient["age"] for patient in patients) / len(patients)
print("Average age:", average_age)
````

# 8. Lambda functions

A lambda is a small anonymous function.

Pattern:

```python
lambda argument: expression
```

Use lambda mainly for short callback functions, such as `key=` in `sorted()`.

Avoid using lambda for complicated business logic. Use a named function instead.

````python
# Sorting patients by risk score
patients_by_risk = sorted(patients, key=lambda patient: patient["risk_score"], reverse=True)

for patient in patients_by_risk:
    print(patient["name"], patient["risk_score"])
````

````python
# Lambda with map/filter examples
scores = [0.72, 0.81, 0.31, 0.89]

percentages = list(map(lambda score: round(score * 100, 1), scores))
high_scores = list(filter(lambda score: score >= 0.75, scores))

print(percentages)
print(high_scores)

# Often, comprehensions are more readable than map/filter + lambda
percentages_v2 = [round(score * 100, 1) for score in scores]
high_scores_v2 = [score for score in scores if score >= 0.75]

print(percentages_v2)
print(high_scores_v2)
````

# 9. Error handling

Use exceptions to handle expected failure cases cleanly.

Common exception patterns:

- `try` / `except`
- `else` when no exception happened
- `finally` for cleanup
- raising your own errors with `raise ValueError(...)`

````python
def parse_blood_pressure(bp: str) -> tuple[int, int]:
    """Parse blood pressure string like '130/85' into systolic and diastolic integers."""
    try:
        systolic_text, diastolic_text = bp.split("/")
        return int(systolic_text), int(diastolic_text)
    except ValueError as exc:
        raise ValueError(f"Invalid blood pressure format: {bp!r}") from exc

print(parse_blood_pressure("130/85"))

try:
    parse_blood_pressure("bad data")
except ValueError as error:
    print("Handled error:", error)
````

# 11. Type hints, docstrings, and clean interfaces

Type hints make code easier to read, review, test, and connect with frameworks like FastAPI.

Examples:

```python
name: str
age: int
scores: list[float]
patient: dict[str, str | int | float | bool]
```

Docstrings explain what a function does, especially when business logic is involved.

````python
from typing import Any

PatientRecord = dict[str, Any]


def summarize_patients(patient_records: list[PatientRecord]) -> dict[str, Any]:
    """Return basic summary statistics for patient records."""
    if not patient_records:
        return {
            "total_patients": 0,
            "average_age": None,
            "active_patients": 0,
            "unique_conditions": [],
        }

    return {
        "total_patients": len(patient_records),
        "average_age": round(sum(p["age"] for p in patient_records) / len(patient_records), 2),
        "active_patients": sum(1 for p in patient_records if p["active"]),
        "unique_conditions": sorted({p["condition"] for p in patient_records}),
    }

summary = summarize_patients(patients)
pprint(summary)
````

# 12. Dataclasses

A `dataclass` is useful when your data has a stable shape and you want cleaner objects than raw dictionaries.

In backend work, you will also see Pydantic models in FastAPI and Django models in Django. Dataclasses are a good Python-native stepping stone.

````python
from dataclasses import dataclass

@dataclass
class Patient:
    patient_id: str
    name: str
    age: int
    condition: str
    visits: int
    risk_score: float
    active: bool = True

    def risk_label(self) -> str:
        if self.risk_score >= 0.75:
            return "high"
        if self.risk_score >= 0.50:
            return "medium"
        return "low"

patient_obj = Patient(
    patient_id="P005",
    name="Hina Shah",
    age=39,
    condition="hypertension",
    visits=3,
    risk_score=0.66,
)

print(patient_obj)
print(patient_obj.risk_label())
````

# 13. Python style standards

Professional Python code should follow common style conventions.

## Naming

- Variables/functions: `snake_case`
- Classes: `PascalCase`
- Constants: `UPPER_SNAKE_CASE`
- Private helper convention: `_helper_function`

## Formatting

- 4 spaces per indentation level
- Keep lines readable
- Use blank lines between logical sections
- Prefer clear names over clever short names
- Use f-strings for readable string formatting

## Function design

Good:

```python
def calculate_risk_label(score: float) -> str:
    ...
```

Bad:

```python
def crl(x):
    ...
```

## Comments

Comments should explain **why**, not repeat obvious code.

````python
# Good constants and naming
HIGH_RISK_THRESHOLD = 0.75
MEDIUM_RISK_THRESHOLD = 0.50


def classify_patient_risk(score: float) -> str:
    """Classify patient risk using shared project thresholds."""
    if score >= HIGH_RISK_THRESHOLD:
        return "high"
    if score >= MEDIUM_RISK_THRESHOLD:
        return "medium"
    return "low"

print([classify_patient_risk(patient["risk_score"]) for patient in patients])
````
